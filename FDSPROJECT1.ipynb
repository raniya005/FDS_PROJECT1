```python
df = df[df['Billing Amount'] >= 0]
```

```python
df.sort_values(by = 'Age')
```

```python
correlation_matrix = df.corr(numeric_only = True)
```

```python
correlation_matrix
```

```python
#visualization
#1 Heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
df.groupby(['Medical Condition', 'Test Results']).size().unstack().plot(kind='bar')
plt.title("Medical Conditions by Test Result")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
avg = df.groupby('Medical Condition')['Billing Amount'].mean()
plt.plot(avg, marker='o', color = 'blue')
plt.title('Average billing vs Medical condition')
plt.xlabel ('Medical condition')
plt.ylabel('Avg billing amount')
plt.show()
```
